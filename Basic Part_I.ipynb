#31. Write a Python program to compute the greatest common divisor (GCD) of two positive integers.

```python
def gcd(m,n):
    for i in range(1,min(m,n)+1):
        if m%i==0 and n%i==0:
            mrcf=i
    return mrcf
m,n=int(input()),int(input())
print(gcd(m,n))
```

#32. Write a Python program to get the least common multiple (LCM) of two positive integers.

```python
def lcm(a,b):
  if a>b:
    greater=a
  else:
    greater=b
  while True:
    if greater%a==0 and greater%b==0:
      lcm=greater
      break
    greater+=1
  return lcm
lcm(8,26)
```

#60. Write a Python program to calculate the hypotenuse of a right angled triangle.

#61. Write a Python program to convert the distance (in feet) to inches, yards, and miles.

```python
def convertor(distance):
  option=int(input(("""
  Choose any one of the following:-
  1. feet to inches
  2. feet to yards
  3. feet to miles
  """)))
  if option==1:
    print(distance,"feet =",distance*12,"inches")
  elif option==2:
    print(distance,"feet =",distance*0.333333,"yards")
  elif option==3:
    print(distance,"feet =",distance*0.00018939375,"miles")
  else:
    print("Invalid Option")
distance=float(input("Enter distance ="))
convertor(distance)
```

66. Write a Python program to calculate body mass index.

```python
def bmi():
    w=float(input("Enter person's weight in kg ="))
    h=float(input("Enter person's height in centimetres ="))
    h=h*0.01
    bmi=w/(h**2)
    if bmi>=25.0:
        return "Overweight"
    elif bmi>18.5 or bmi<24.9:
        return "Healthy"
    else:
        return "Underweight"
```
